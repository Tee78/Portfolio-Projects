```python


sales_data = [
    {"item": "Coffee", "quantity": 10, "price": 500},
    {"item": "Tea", "quantity": 8, "price": 300},
    {"item": "Sandwich", "quantity": 5, "price": 1200}
]

items = [record["item"] for record in sales_data]
revenue = [record["quantity"] * record["price"] for record in sales_data]
```

```python
# Visualising revenue distribution across products for comparison


plt.figure(figsize=(8, 5))
plt.bar(items, revenue)
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.yticks(range(0, max(revenue)+1, 1000))
plt.show()
```
